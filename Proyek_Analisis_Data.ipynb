## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana distribusi total pembayaran?

```python
import pandas as pd
import matplotlib.pyplot as plt

order_payments_dataset = pd.read_csv('order_payments_dataset.csv')

order_payments_dataset['payment_type'] = order_payments_dataset['payment_type'].replace({'debit_card': 'others', 'not_defined': 'others'})

percentage_total_payment = (order_payments_dataset['payment_type'].value_counts(normalize=True)* 100).sort_index()

colors = plt.cm.Paired.colors

plt.pie(
    percentage_total_payment,
    labels=percentage_total_payment.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.4),
    textprops=dict(size=7),
)

plt.title('Distribusi Total Pembayaran')
plt.axis('equal')
plt.show()
```

### Pertanyaan 2: Bagaiamana tren penjualan peningkatan atau penurunan pembayaran dari waktu ke waktu?

```python
import matplotlib.pyplot as plt

total_payment_per_type = order_payments_dataset.groupby('payment_type').sum()


plt.figure(figsize=(10,6))
plt.plot(total_payment_per_type.index, total_payment_per_type['payment_value'], marker='o', color='skyblue', linestyle='-')
plt.title('Tren Penjualan dan Pembayaran dari Waktu ke Waktu')
plt.xlabel('Jenis Pembayaran')
plt.ylabel('Total Pembayaran')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

- Conclution pertanyaan 1 : Bagaimana distribusi total pembayaran?

Berdasarkan pie chart distribusi total pembayaran diatas dapat diketahui bahwa, penggunaan credit card paling banyak yaitu dengan jumlah sebesar 73,9%. - Conclution pertanyaan 2 : Bagaiamana tren penjualan peningkatan atau penurunan pembayaran dari waktu ke waktu?

Penjualan dan pembayaran menunjukkan tren positif secara keseluruhan, dengan peningkatan dari waktu ke waktu. Credit card adalah metode pembayaran yang paling populer, diikuti oleh other, boleto, dan vocher. Tren penjualan dan pembayaran mengalami fluktuasi, dengan beberapa periode kenaikan dan penurunan.
